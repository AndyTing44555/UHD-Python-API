#### Application Notes
This code is intented to generate and output a waveform from USRP B210

```python
import uhd
import numpy as np
import os
```

```python
### change directory to the working folder (.../code)
code_dir = os.getcwd()
parent_dir = os.path.dirname(code_dir)
figs_dir = os.path.join(parent_dir, 'figures')
funs_dir = os.path.join(parent_dir, 'functions')
data_dir = os.path.join(parent_dir, 'data')

if not os.path.exists(figs_dir):
    os.makedirs(figs_dir)
if not os.path.exists(funs_dir):
    os.makedirs(funs_dir)
if not os.path.exists(data_dir):
    os.makedirs(data_dir)
       
print(parent_dir)
```

```python
### Testinng Connection of USRP B210 (type is B200)
try:
    my_usrp= uhd.usrp.MultiUSRP("type=b200")  # no device present
    # print("USRP device created:", my_usrp.get_pp_string())
    print("USRP device created")
except Exception as e_usrp:
    print("Failure (no device):", type(e_usrp).__name__)
    print(e_usrp)
```

```python
### Sapling parameters
mod_freq = int(1e5)  # modulation frequency
# center_freq = int(50e6)
center_freq = int(1e6)
samp_rate = int(16e6)

t_array = np.arange(0, 1, 1/samp_rate)
```

```python
# my_usrp = uhd.usrp.MultiUSRP()
# samps = 0.1*np.random.randn(10000) + 0.1*1j*np.random.randn(10000) # create random signal
samps = np.sin(2*np.pi*mod_freq*t_array, dtype=np.complex64)
duration = 10 # seconds
gain = 20 # [dB] start low then work your way up
my_usrp.send_waveform(samps, duration, center_freq, samp_rate, [0], gain)
```
